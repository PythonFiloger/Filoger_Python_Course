### break, continue

```python
for i in range(6):
  score = input(f"score_{i+1}: ")
  if score == "exit":
    break
  print(float(score))
```

```python
for number in range(7,10):
  print("number: ", number)

for number in range(11,14):
  print("number: ", number)
```

```python
for number in range(7,14):
  print(number, end=",")
  print("python", end=",")
  if number == 8:
    continue
  print("java", end=",")
  print("c++", end=",")
```

```python
score1 = float(input("your score: "))
max_score = score1
sum_score = score1
for i in range(3):
  score = float(input("score: "))
  sum_score += score
  if score > max_score:
    max_score = score
print(f"max scores: {max_score}")
avg = sum_score / 4
print(f"avg: {avg}")
```
